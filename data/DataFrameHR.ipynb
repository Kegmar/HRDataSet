```python
import pandas as pd

data = {
    'Name': ['Dairis', 'Olga', 'Edgars', 'Svetlana 1', 'Svetlana 2', 'Svetlana 3', 'Vladimirs', 'Elena', 'Vadim', 'Natasha', 'Natasha 2', 'Tatjana', 'Irina', 'Maksims'],
    'Salary': [1300, 950, 850, 750, 650, 850, 700, 450, 600, 700, 750, 780, 780, 760],
    'Age': [35, 45, 36, 55, 60, 60, 57, 36, 31, 45, 59, 46, 55, 26]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Calculate correlation
correlation = df['Salary'].corr(df['Age'])

print("Correlation coefficient between age and salary", correlation)
```

```python
import matplotlib.pyplot as plt
# Histogram of salaries
plt.hist(df['Salary'], bins=10, edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.title('Distribution of Salaries')
plt.show()
```

```python
# Bar chart of ages
plt.bar(df['Name'], df['Age'])
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Ages of Employees')
plt.xticks(rotation =45)
plt.show()
```

```python
# Scatter plot of salary vs. age
plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary vs. age')
plt.show()
```
